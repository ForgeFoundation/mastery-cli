Create a Binomial distribution with:
  - [ ] 10 numbers 
  - [ ] 0.5 probability
  - [ ] Name it `Binomial Distribution (n=10, p=0.5)`
  - [ ] Show an Barchart of it


![](./../img/2023-10-23-14-21-21.png)

```python
n = 10
p = 0.5

x = range(n + 2)
y = stats.binom.pmf(x, n, p)

plt.bar(x, y)
plt.title('Analysis of Binomial Distribution')
plt.show()
```

```python
## binomial

# a binomial distribution is the probability of K heads in N coin tosses,
# given a probability of p heads (e.g., .5 is a fainr coin).

n = 10 # number on coin tosses
p = .5 # probability of heads

x = range(n+2)
bindist = stats.binom.pmf(x,n,p)

plt.bar(x,bindist)
plt.title('Binomial distribution (n=%s, p=%g)'%(n,p))
plt.show()
```

- [ ] Create an F distribution with:
  - [ ] 1001 random numbers
  - [ ] 5 degreees of freedom
  - [ ] Of Density 100
  - [ ] Name it `F (5, 100) Distribution`
  - [ ] Show a linear plot

![](./../img/2023-10-23-14-26-15.png)

```python
num_df = 5
den_df = 50
x = np.linspace(0, 10, 100)
y = stats.f.pdf(x, num_df, den_df)
plt.plot(x, y)
plt.show()
```

```python
## F

# parameters
num_df = 5   # numerator degrees of freedom
den_df = 100 # denominator df

# values to evaluate 
x = np.linspace(0,10,100)

# the distribution
fdist = stats.f.pdf(x,num_df,den_df)

plt.plot(x,fdist)
plt.title(f'F({num_df},{den_df}) distribution')
plt.xlabel('F value')
plt.show()
```
